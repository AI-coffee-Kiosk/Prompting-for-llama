````python
def detect_action_type(user_input):
    # Define some simple keyword-based rules to infer the action type
    if any(keyword in user_input for keyword in ["취소", "삭제", "취소해주세요"]):
        return "cancel_order"
    elif any(keyword in user_input for keyword in ["바꿔", "대신", "변경"]):
        return "modify_order"
    elif any(keyword in user_input for keyword in ["추가", "넣어", "더해"]):
        return "add_options"
    elif any(keyword in user_input for keyword in ["빼", "제거"]):
        return "remove_options"
    elif any(keyword in user_input for keyword in ["완료", "끝내", "계산"]):
        return "finish_order"
    elif any(keyword in user_input for keyword in ["추천", "추천하는", "추천해줘"]):
        return "recommend_coffee"
    else:
        return "order_coffee"  # Default to order if no other keywords match
````

````python
# Initialize global conversation history and order state
conversation_history = []
current_order = {
    "action": "create_order",
    "orders": []
}

def detect_action_type(user_input):
    """
    Placeholder function to detect action type based on input.
    For example, if the user says 'cancel', it returns 'cancel_order'.
    """
    # Implement logic to detect the action based on the input
    if "바꿔" in user_input or "대신" in user_input:
        return "modify_order"
    elif "취소" in user_input:
        return "cancel_order"
    elif "추가" in user_input or "빼" in user_input:
        return "add_or_remove_options"
    elif "완료" in user_input or "확인" in user_input:
        return "finish_order"
    elif "추천" in user_input:
        return "recommend_coffee"
    else:
        return "order_coffee"

def update_order_state(user_input):
    global current_order, conversation_history

    # Update conversation history with the new input
    conversation_history.append(user_input)

    # Detect action type based on the new input
    action_type = detect_action_type(user_input)

    # Modify the current order based on detected action
    if action_type == "order_coffee":
        new_order = {
            "drink": "아메리카노",  # You will dynamically detect the drink type
            "size": "미디움",  # Default size
            "temperature": "핫",  # Default temperature
            "quantity": 1,  # Default quantity
            "add_ons": [],
            "extra_shots": 0
        }
        current_order["orders"].append(new_order)

    elif action_type == "modify_order":
        # Example modification logic (modify the drink type)
        for order in current_order["orders"]:
            if order["drink"] == "아메리카노":  # Detect what to modify
                order["drink"] = "라떼"

    elif action_type == "add_or_remove_options":
        # Example logic for adding/removing options
        for order in current_order["orders"]:
            if "추가" in user_input:
                order["add_ons"].append("휘핑크림")
            elif "빼" in user_input:
                if "휘핑크림" in order["add_ons"]:
                    order["add_ons"].remove("휘핑크림")

    elif action_type == "cancel_order":
        current_order["orders"] = []  # Cancel the order

    elif action_type == "finish_order":
        return generate_order_summary(current_order)  # Generate the final order summary

    elif action_type == "recommend_coffee":
        return generate_coffee_recommendation()

    # Generate a new prompt with updated context
    prompt = generate_prompt_based_on_history(conversation_history, current_order)

    # Send the prompt to LLaMA and get the response
    model_response = get_model_response(prompt)

    return model_response

def generate_prompt_based_on_history(conversation_history, current_order):
    history_text = " ".join(conversation_history)

    # Example of a dynamic prompt that includes conversation history and current order state
    prompt = f"""
    You are a virtual coffee kiosk. The following is the conversation history with the customer:
    {history_text}

    Current order state:
    {current_order}

    The customer has provided the input: "{conversation_history[-1]}"

    Based on this input, update the order and respond with a natural language confirmation and a JSON file.
    """

    return prompt

def generate_order_summary(order):
    # Create a natural language summary of the current order
    summary = []
    for item in order["orders"]:
        item_summary = f"{item['size']} {item['temperature']} {item['drink']} {item['quantity']} 잔"
        if item['add_ons']:
            item_summary += f" 추가 옵션: {', '.join(item['add_ons'])}"
        summary.append(item_summary)

    return f"주문이 완료되었습니다. 주문하신 음료는 {', '.join(summary)}입니다. 감사합니다."

def generate_coffee_recommendation():
    # Example coffee recommendation response
    return "저희 매장에서 추천하는 커피는 아메리카노, 카페라떼, 카푸치노, 에스프레소입니다. 어떤 것을 원하시나요?"

# Mock function for model response (replace with actual LLaMA interaction)
def get_model_response(prompt):
    # This function would call LLaMA with the prompt and return the generated response
    print(f"Prompt sent to model:\n{prompt}")
    return "주문이 완료되었습니다. 손님의 주문은 라떼 미디움 한 잔입니다."

# Example usage
user_input1 = "아메리카노 한 잔 주세요."
user_input2 = "휘핑크림 추가해줘."
user_input3 = "라떼로 바꿔줘."
user_input4 = "주문 완료할게요."

# Process each input sequentially
response1 = update_order_state(user_input1)
print(response1)

response2 = update_order_state(user_input2)
print(response2)

response3 = update_order_state(user_input3)
print(response3)

response4 = update_order_state(user_input4)  # Final order confirmation
print(response4)
````

````python
from transformers import pipeline

# Initialize model
model = pipeline('text-generation', model='Bllossom/llama-3.2-Korean-Bllossom-3B', device=0)

# Initial values for previous response
previous_response = {
    "user_input": "",
    "natural_response": "",
    "json_response": "{}"
}
# Function to update and generate prompt with feedback
def generate_prompt_with_feedback(user_input, previous_response):
    prompt = f"""
    You are operating a virtual coffee kiosk that receives speech-to-text (STT) inputs from customers placing coffee orders.
    Your role is to understand and process these inputs, respond naturally in Korean, and generate a structured JSON file with the correct details for backend processing.

    **Key Requirements:**
    - **Menu Items:** The kiosk offers the following drinks: 아메리카노, 라떼, 카푸치노, 카페모카, 바닐라라떼, 에스프레소, 카라멜마끼아또, 허브티, 홍차, 초콜릿라떼, 레몬에이드, 복숭아아이스티, 딸기스무디, 망고스무디, 키위주스, 토마토주스.
    - **Default Values:**
      - Default size is **미디움** (medium).
      - Default temperature is **핫** (hot).
    - **Misspelled or Short Names:** If the customer uses a shorthand or misspelling of a drink name (e.g., 아아 for 아이스 아메리카노 or 뜨아 for 핫 아메리카노), you must understand and interpret the request correctly.
    - **Out of Menu Items:** If the customer requests an item that is not on the menu, you must politely respond that the item is unavailable and suggest they try another drink from the menu. No action should be taken for invalid items, and no JSON should be generated in such cases.
    - **Multiple Requests:** The customer might give multiple requests in one interaction. You must be able to process and remember each step until the customer confirms the order.
    - **Customer Confirmation:** The order should be finalized only after the customer confirms it. Until then, you should continue to remember their requests.
    - **Functions:** You must support the following actions:
      - Create an order
      - Modify an existing order (e.g., changing a drink or adding/removing an option)
      - Cancel an order
      - Recommend coffee or suggest closest items if what they requested is not available.
      - Confirm the order.

    **Example Scenarios:**

    1. **Basic Coffee Order:**
       - Input: "라지 아이스 아메리카노 두 잔 주세요."
       - Natural Language Response: "주문이 완료되었습니다. 손님의 주문은 라지 아이스 아메리카노 두 잔입니다."
       - JSON Output:
       {{
           "action": "create_order",
           "drink": "아메리카노",
           "size": "라지",
           "temperature": "아이스",
           "quantity": 2,
           "add_ons": [],
           "extra_shots": 0
       }}

    2. **Adding Options to a Coffee Order:**
       - Input: "휘핑크림 추가 해주세요."
       - Natural Language Response: "휘핑크림이 추가되었습니다."
       - JSON Output:
       {{
           "action": "modify_order",
           "drink": "아메리카노",
           "size": "미디움",
           "temperature": "핫",
           "quantity": 1,
           "add_ons": ["휘핑크림"],
           "extra_shots": 0
       }}

    3. **Modifying an Existing Order:**
       - Input: "주문한거 라떼로 바꿔줘."
       - Natural Language Response: "주문이 아메리카노에서 라떼로 변경되었습니다."
       - JSON Output:
       {{
           "action": "modify_order",
           "old_drink": "아메리카노",
           "new_drink": "라떼",
           "size": "미디움",
           "temperature": "핫",
           "quantity": 1,
           "add_ons": [],
           "extra_shots": 0
       }}

    4. **Misspelled or Short Drink Name:**
       - Input: "아아 한 잔 주세요."
       - Natural Language Response: "주문이 완료되었습니다. 손님의 주문은 아이스 아메리카노 한 잔입니다."
       - JSON Output:
       {{
           "action": "create_order",
           "drink": "아메리카노",
           "size": "미디움",
           "temperature": "아이스",
           "quantity": 1,
           "add_ons": [],
           "extra_shots": 0
       }}

    5. **Item Not on Menu:**
       - Input: "초코라떼 주세요."
       - Natural Language Response: "죄송합니다, 초코라떼는 메뉴에 없습니다. 대신 초콜릿라떼를 추천드립니다."
       - JSON Output:
       {{
           "action": "recommend_closest_item",
           "requested_item": "초코라떼",
           "recommended_item": "초콜릿라떼"
       }}

    6. **Order Confirmation:**
       - Input: "주문 완료할게요."
       - Natural Language Response: "주문이 완료되었습니다. 결제는 카드리더기를 사용해주세요. 손님의 주문은 아메리카노 미디움 한 잔입니다. 감사합니다."
       - JSON Output:
       {{
           "action": "complete_order",
           "drink": "아메리카노",
           "size": "미디움",
           "temperature": "핫",
           "quantity": 1,
           "add_ons": [],
           "extra_shots": 0
       }}

    Process the customer's new input, taking into account any prior information as needed.
    Generate a response with:
    1. A natural language response confirming the action taken.
    2. A structured JSON file with updated order details or other relevant information.
    **Key Response Requirements:**
    1. **Response Structure:** Each response must consist of exactly two parts:
      - A natural language confirmation response to the customer.
      - A structured JSON output summarizing the order details.

    2. **Current Customer Focus:** Provide responses only relevant to the current customer's order without referring to any prior context unless prompted.
    3. **Avoid Extra Responses:** Do not generate any additional explanations or comments outside the two required response parts for each customer's new input only 1 response should be given.
 Response Structure Requirements:

Each response must contain exactly two parts:

Natural Language Response: A clear confirmation or clarification based solely on the customer’s new input.
Structured JSON Output: Summarize only the current order’s relevant details, formatted strictly as JSON with no extra comments or explanations.
Avoid Extra Explanations: Do not provide additional information, examples, or extra comments beyond these two required parts.

Focus on Current Customer Input:

Each response should reflect only the customer’s current order without referencing any prior example or extraneous information. Use exactly the details of the latest request, incorporating prior context only if the customer directly modifies or adds to their previous order.
Guidelines for Handling Prior Context:

Only use prior interaction details to modify or add to an existing order if the customer explicitly directs you to do so (e.g., "Add whipped cream" or "Change my drink to a latte").
Otherwise, disregard past interactions unless the customer makes clear that they are continuing a prior order modification.
Example Corrections and Feedback:

Incorrect Responses:

Including unnecessary information or examples from past responses.
Generating multiple versions of responses for a single customer input.
Correct Responses:
  **Examples:**

    Customer: "아이스 아메리카노 한잔 주세요."
    Correct Natural Language Response: "주문이 완료되었습니다. 손님의 주문은 아이스 아메리카노 한 잔입니다."
    Correct JSON Output:
      {{
           "action": "create_order",
           "drink": "아메리카노",
           "size": "미디움",
           "temperature": "아이스",
           "quantity": 1,
           "add_ons": [],
           "extra_shots": 0
       }}

    Response Quality Assurance:

    Each response should be concise, accurate to the customer’s request, and should appear as if a dedicated coffee kiosk is processing the order.
    Use a clear, direct style in the natural language response and maintain precision in the JSON structure to match the exact details requested by the customer.
    Guiding Principle: Treat each input as if it is a fresh interaction, handling only the most recent details unless the customer specifically indicates they are modifying a prior request. Respond with the precision and consistency expected from a dedicated automated ordering kiosk.
    Enhanced Natural Language Response:

  In the natural language response, summarize each item in the order with a clear breakdown of the specific drinks, quantities, sizes, temperature, and any add-ons requested. Ensure this summary is easy to understand and confirms all details explicitly for the customer.

  Example Natural Language Response:

  For the customer input: "아이스 아메리카노 4잔이랑 카페라떼 라지로 2잔, 아이스 카라멜마끼아또 엑스라지로 휘핑크림 추가해서 3잔 주세요," a suitable response would be:

  주문하신 내용은 다음과 같습니다:
  - 아이스 아메리카노 4잔
  - 라지 카페라떼 2잔 (기본 핫 옵션)
  - 아이스 카라멜마끼아또 엑스라지 3잔 (휘핑크림 추가)
  주문을 확인해주시고, 추가나 수정이 필요하면 말씀해주세요.
  Detailed JSON Structure:
  {{
      "action": "create_order",
      "order_items": [
          {{
              "drink": "아메리카노",
              "size": "미디움",
              "temperature": "아이스",
              "quantity": 4,
              "add_ons": [],
              "extra_shots": 0
          }},
          {{
              "drink": "카페라떼",
              "size": "라지",
              "temperature": "핫",
              "quantity": 2,
              "add_ons": [],
              "extra_shots": 0
          }},
          {{
              "drink": "카라멜마끼아또",
              "size": "엑스라지",
              "temperature": "아이스",
              "quantity": 3,
              "add_ons": ["휘핑크림"],
              "extra_shots": 0
          }}
      ]
    }}
    Concise Response Format: When summarizing the order in the natural language response, combine identical items with add-ons into a single line. For example, instead of listing separate lines like "- 아이스 아메리카노 두잔" and "- 아이스 아메리카노에 샷 추가," respond with "- 아이스 아메리카노 2잔 (1샷 추가)" to create a clear and compact summary.
    Example for Customer Request: If the customer requests "아이스 아메리카노에 샷 추가해서 두잔 주세요", the response should be:

    Natural Language Response: "주문하신 내용은 다음과 같습니다:
     -아이스 아메리카노 2잔 (1샷 추가)"
    JSON Output:
    {{
      "action": "create_order",
      "order_items": [
          {{
              "drink": "아메리카노",
              "size": "미디움",
              "temperature": "아이스",
              "quantity": 2,
              "add_ons": ["샷"],
              "extra_shots": 1
          }}
      ]
    }}

    **Customer's New Input:** "{user_input}"

    **Previous Interaction**:
    - Customer's Last Request: "{previous_response['user_input']}"
    - Last Natural Language Response: "{previous_response['natural_response']}"
    - Last JSON Response: {previous_response['json_response']}

    Based on information above what would be the correct natural language response: and correct json output for the **Customer's New Input:** "{user_input}".
    """
    return prompt
# Function to process the model response
def get_model_response(prompt):
    response = model(prompt, max_new_tokens=200, truncation=True)[0]['generated_text']
    try:
        natural_response, json_output = response.split("JSON Output:", 1)
    except ValueError:
        natural_response, json_output = "Error: Invalid response format", "{}"
    return natural_response.strip(), json_output.strip()

# Main interaction loop
while True:
    user_input = input("Customer: ")
    prompt = generate_prompt_with_feedback(user_input, previous_response)
    natural_response, json_output = get_model_response(prompt)

    # Print the feedback-enforced responses
    print(f"Natural Language Response: {natural_response}")
    print(f"JSON Output: {json_output}")

    # Update previous response for feedback context
    previous_response["user_input"] = user_input
    previous_response["natural_response"] = natural_response
    previous_response["json_response"] = json_output

    # Break if customer confirms the order
    if "주문 완료" in user_input:
        break
````
