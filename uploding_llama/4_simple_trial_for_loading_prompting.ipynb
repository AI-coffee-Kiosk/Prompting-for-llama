```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
import warnings

# Suppress unnecessary warnings
warnings.filterwarnings("ignore")  # Suppress all warnings

# Initialize model and tokenizer with optimizations
model_id = "Bllossom/llama-3.2-Korean-Bllossom-3B"
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id, torch_dtype=torch.float16, device_map="auto"
)

# Set padding and end tokens
tokenizer.pad_token = tokenizer.eos_token

# Define the input prompt
input_text = "머신러닝을 공부하려면 어떻게 시작해야 하나요?"

# Tokenize input with padding and truncation
inputs = tokenizer(input_text, return_tensors="pt", padding=True, truncation=True)
device = "cuda" if torch.cuda.is_available() else "cpu"
inputs = inputs.to(device)

# Dynamic `max_new_tokens` based on input length, up to 512 for faster output
max_new_tokens = min(512, len(inputs.input_ids[0]) + 100)

# Define optimized generation parameters
generation_params = {
    "input_ids": inputs.input_ids,
    "attention_mask": inputs.attention_mask,
    "do_sample": True,             # Enables sampling for varied and creative responses
    "temperature": 0.7,            # Temperature for response creativity
    "top_p": 0.9,                  # Nucleus sampling to allow diverse responses
    "repetition_penalty": 1.1,     # Reduces repetitiveness in responses
    "pad_token_id": tokenizer.eos_token_id,
    "eos_token_id": tokenizer.eos_token_id,
    "max_new_tokens": 10000          # Adjusts dynamically as the model generates
}

# Generate response and decode output
with torch.no_grad():
    output = model.generate(**generation_params)
    response = tokenizer.decode(output[0], skip_special_tokens=True)

print("Response:", response)
```
